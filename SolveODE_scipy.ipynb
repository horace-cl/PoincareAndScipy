## $x'' + 5x' + x^5 - 3x^3 + x = 10 \sin(t)$

Hay que convertirlas a ecuaciones de primer orden!
<br>

#### $x' = x_1$
#### $x_1' = -5x_1-x^5+3x^3-x+10\sin(t)$

```python
def diferentialEq(w, t):
    #La ecuaciones se escriben como entradas de una `tupla`
    #El orden importa!
    #w contiene x1 y x
    return (w[1], -5*w[1]-w[0]**5+3*w[0]**3-w[0]+10*np.cos(t))
```

```python
#Condiciones Iniciales
x0 = [0,0]
#Densidad de puntos en la variable independiente
t = np.linspace(-10, 10,1000)
```

```python
#Resuelve el sistema de ecuaciones de 1er orden
res = integrate.odeint(diferentialEq, x0, t)
```

```python
#La columna 0 corresponde a x y la columna 1 a x1 = x'
res_x = res[:,0]
res_xp = res[:,1]
```

```python
plt.figure(figsize=(10,10))
plt.plot(res_x, res_xp)
plt.xlabel('$x$')
plt.ylabel("$x'$")
#plt.set_aspect('equal')
```

```python
plt.figure(figsize=(10,10))
plt.plot(t, res_xp)
plt.xlabel('$t$')
plt.ylabel("$x'$")
#plt.set_aspect('equal')
```

```python
plt.figure(figsize=(10,10))
plt.plot(t, res_x)
plt.xlabel('$t$')
plt.ylabel("$x$")
```
